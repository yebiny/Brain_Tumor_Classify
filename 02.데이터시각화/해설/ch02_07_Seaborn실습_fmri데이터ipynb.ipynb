### (3) Relational plot

![](https://seaborn.pydata.org/_images/function_overview_8_0.png)

```
seaborn.relplot(x=None, y=None, 
                hue=None, 
                size=None, 
                style=None, 
                data=None, 
                row=None, col=None, 
                col_wrap=None,
                row_order=None, col_order=None, 
                palette=None, 
                hue_order=None, hue_norm=None, 
                sizes=None, size_order=None, size_norm=None,
                markers=None, dashes=None, style_order=None,
                legend='auto', # “auto”, “brief”, “full”, or False
                kind='scatter', # 'scatter', 'line'
                height=5, aspect=1, 
                facet_kws=None, 
                units=None)
```

````python
sns.relplot(data=fmri, x='timepoint', y='signal', kind='line')
plt.title('fmri line plot')
plt.show()
````

````python
sns.relplot(data=fmri, x='timepoint', y='signal', kind='line',
            hue='event', style='region', size='region', 
            markers=False, dashes=False)
````

````python
sns.relplot(data=fmri, x='timepoint', y='signal', kind='line',
            hue='subject', col='event', row='region',
            palette='Spectral',
            hue_order=[f's{i}' for i in range(14)])
plt.show()
````

````python
g = sns.relplot(x="timepoint", y="signal", hue="event", style="event",
                col="subject", col_wrap=5,
                col_order=[f's{i}' for i in range(14)],
                height=3, aspect=.75, linewidth=2.5,
                kind="line", data=fmri.query("region == 'frontal'"),
                )
````
